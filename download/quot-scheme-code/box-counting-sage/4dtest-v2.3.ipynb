# Compare with CK Conjecture 3.12

$$\sum q^n\int_{[X^{[n]}]^{\text{vir}}}c_n(L^{[n]})=M(-q)^{-\frac{(w_1t_1+w_2t_2+w_3t_3)(t_1+t_2)(t_1+t_3)(t_2+t_3)}{t_1t_2t_3(t_1+t_2+t_3)}}$$
$$=\text{Exp}\left(\frac{(w_1t_1+w_2t_2+w_3t_3)(t_1+t_2)(t_1+t_3)(t_2+t_3)q}{t_1t_2t_3(t_1+t_2+t_3)(1+q)^2}\right)$$
$$=\text{Exp}\left(\frac{(w_1t_1+w_2t_2+w_3t_3)(t_1+t_2)(t_1+t_3)(t_2+t_3)}{t_1t_2t_3(t_1+t_2+t_3)}\sum_{m=1}^\infty (-1)^{m+1}mq^m\right)$$
$$=\text{exp}\left(\sum_{n=1}^\infty\frac{(w_1t_1+w_2t_2+w_3t_3)(t_1+t_2)(t_1+t_3)(t_2+t_3)}{nt_1t_2t_3(t_1+t_2+t_3)}\sum_{m=1}^\infty (-1)^{mn+1}mq^{mn}\right)$$
$$=\sum_{k=0}^\infty\frac{1}{k!}\left(\sum_{n=1}^\infty\frac{(w_1t_1+w_2t_2+w_3t_3)(t_1+t_2)(t_1+t_3)(t_2+t_3)}{nt_1t_2t_3(t_1+t_2+t_3)}\sum_{m=1}^\infty (-1)^{mn+1}mq^{mn}\right)^k$$

When the power on $q$ is $1$, we have coefficient:
$$\frac{(w_1t_1+w_2t_2+w_3t_3)(t_1+t_2)(t_1+t_3)(t_2+t_3)}{t_1t_2t_3(t_1+t_2+t_3)}$$
Indeed, we check:

$$+[k=1,n=1,m=2] -\frac{2(w_1t_1+w_2t_2+w_3t_3)(t_1+t_2)(t_1+t_3)(t_2+t_3)}{t_1t_2t_3(t_1+t_2+t_3)} $$
